```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
data = pd.read_csv('E:\\Skt\\DA_Prep\\Python\\householdtask3.csv')
```

```python
data.head(20)
```

```python
# scatter plot - income vs year
plt.scatter(data['year'],data['income'])

# adding Title for plot
plt.title("Scatter Plot")

# Setting labels
plt.xlabel('year')
plt.ylabel('income')

# showing the plot
plt.show()
```

```python
# line chart - income vs expenditure
plt.plot(data['expenditure'])
plt.plot(data['income'])

# adding Title for the chart
plt.title("Line Chart")

# Setting up labels
# Setting labels
plt.xlabel('expenditure')
plt.ylabel('income')

# showing the plot
plt.show()
```

```python
# Bar chart - income vs year
plt.bar(data['year'],data['income'])

# adding Title for plot
plt.title("Bar Chart")

# Setting labels
plt.xlabel('year')
plt.ylabel('income')

# showing the plot
plt.show()
```

```python
# Histogram
plt.hist(data['expenditure'])

# adding title
plt.title("Histogram")

#showing the plot
plt.show()
```

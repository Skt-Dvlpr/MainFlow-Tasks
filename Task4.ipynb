```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data_frame = pd.read_csv('E:\\Skt\\DA_Prep\\Python\\USvideos.csv')
```

```python
data_frame.head()
```

```python
data_frame.shape
```

```python
data_frame = data_frame.drop_duplicates()
data_frame.shape
```

```python
data_frame.describe()
```

```python
data_frame.info()
```

```python
remove_column = ['thumbnail_link','description']
data_frame = data_frame.drop(columns = remove_column)
data_frame.info()
```

```python
import datetime

data_frame['trending_date'] = data_frame["trending_date"].apply(lambda x: datetime.datetime.strptime(x, '%y.%d.%m'))
data_frame.head()
```

```python
data_frame['publish_time'] = pd.to_datetime(data_frame['publish_time'])
data_frame.head()
```

```python
data_frame['publish_month'] = data_frame['publish_time'].dt.month
data_frame['publish_day'] = data_frame['publish_time'].dt.day
data_frame['publish_hour'] = data_frame['publish_time'].dt.hour
data_frame.head(3)
```

```python
print(sorted(data_frame["category_id"].unique()))
```

```python
data_frame['category_name'] = np.nan
data_frame.loc[(data_frame["category_id"] == 1), "category_name"] = 'Film and Animation'
data_frame.loc[(data_frame["category_id"] == 2), "category_name"] = 'Autos and Vehicles'
data_frame.loc[(data_frame["category_id"] == 10), "category_name"] = 'Music'
data_frame.loc[(data_frame["category_id"] == 15), "category_name"] = 'Pets and Animals'
data_frame.loc[(data_frame["category_id"] == 17), "category_name"] = 'Sports'
data_frame.loc[(data_frame["category_id"] == 19), "category_name"] = 'Travel and Events'
data_frame.loc[(data_frame["category_id"] == 20), "category_name"] = 'Gaming'
data_frame.loc[(data_frame["category_id"] == 22), "category_name"] = 'People and Blogs'
data_frame.loc[(data_frame["category_id"] == 23), "category_name"] = 'Comedy'
data_frame.loc[(data_frame["category_id"] == 24), "category_name"] = 'Entertainment'
data_frame.loc[(data_frame["category_id"] == 25), "category_name"] = 'News and Politics'
data_frame.loc[(data_frame["category_id"] == 26), "category_name"] = 'How to and Style'
data_frame.loc[(data_frame["category_id"] == 27), "category_name"] = 'Education'
data_frame.loc[(data_frame["category_id"] == 28), "category_name"] = 'Science and Technology'
data_frame.loc[(data_frame["category_id"] == 29), "category_name"] = 'Non Profit and Activism'
data_frame.loc[(data_frame["category_id"] == 43), "category_name"] = 'Shows'
data_frame.head()
```

```python
data_frame['year'] = data_frame['publish_time'].dt.year
yearly_count = data_frame.groupby('year')['video_id'].count()

#Bar chart
yearly_count.plot(kind='bar', xlabel='Year', ylabel='Total PublishCount', title='Total Publish Video Per Year')
plt.show()
```

```python
#Group by year and sum of views each year
yearly_view = data_frame.groupby('year')['views'].sum()

#Bar chart
yearly_view.plot(kind='bar', xlabel='Year', ylabel='Total Views', title='Total Views Per Year')
plt.xticks(rotation = 0)
plt.tight_layout()

plt.show()
```

```python
#Group by category_name and sum of views in each category
category_view = data_frame.groupby('category_name')['views'].sum().reset_index()

#sorting
top_categories = category_view.sort_values(by='views', ascending=False).head(5)

#Bar chart
plt.bar(top_categories['category_name'], top_categories['views'])
plt.xlabel('Category Name', fontsize=12)
plt.ylabel('Total Views', fontsize=12)
plt.title('Top 5 Categories', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='category_name', data= data_frame, order = data_frame['category_name'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Video Count by Category')
plt.show()
```

```python
videos_per_hour = data_frame['publish_hour'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
sns.barplot(x=videos_per_hour.index, y=videos_per_hour.values, palette='rocket')
plt.title('Number of Videos Published per hour')
plt.xlabel('Hour of the day')
plt.ylabel('Number of Videos')
plt.xticks(rotation = 45)
plt.show()
```

```python
data_frame['publish_time'] = pd.to_datetime(data_frame['publish_time'])
data_frame['publish_date'] = data_frame['publish_time'].dt.date
video_count_by_date = data_frame.groupby('publish_date').size()
plt.figure(figsize=(12,6))
sns.lineplot(data=video_count_by_date)
plt.title("Videos Published Over time")
plt.xlabel("Publish Date")
plt.ylabel("Number of Videoas")
plt.xticks(rotation=45)
plt.show()
```

```python
#scatter plot
sns.scatterplot(data= data_frame, x='views', y='likes')
plt.title("Views V/S Likes")
plt.xlabel('Views')
plt.ylabel("Likes")
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.subplots_adjust(wspace = 0.2, hspace =0.4, top =0.9)
plt.subplot(2,2,1)
g =sns.countplot(x='comments_disabled', data=data_frame)
g.set_title("Comments Disabled", fontsize=16)
plt.subplot(2,2,2)
h =sns.countplot(x='ratings_disabled', data= data_frame)
h.set_title("Rating Disabled", fontsize=16)
plt.subplot(2,2,3)
i= sns.countplot(x='video_error_or_removed', data=data_frame)
i.set_title("Video Error or Removed", fontsize=16)
plt.show()
```
